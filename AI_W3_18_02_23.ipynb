```python
import pandas as pd
dataframe = pd.DataFrame()
dataframe['Name'] = ['Lemon','Apple','Banana']
dataframe['Price'] = [5,8,7]
dataframe['License'] = [False,True,True]
dataframe
```

```python
dataframe.shape
```

```python
df=dataframe
df.info()
```

```python
df.describe()
#mean là stb
#std là độ lệch chuẩn
```

```python
# loc iloc : location
df.loc[(df.Price == 8)]
```

```python
df.loc[(df.Price == 8) | (df.Name == 'Banana')] 
# ' | ' là toán tử or
```

```python
df.loc[(df.Price == 8) | (df.Name == 'Banana') , ['Name','Price']]
```
